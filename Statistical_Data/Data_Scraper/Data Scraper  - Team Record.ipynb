```python
import time
import re
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
```

```python
def work_player_profile(team):
    url = 'https://www.basketball-reference.com/teams/' + team
    res = requests.get(url)
    soup = BeautifulSoup(res.text)

    print(f"Current team: {team}")

    data_dict = defaultdict(list)

    win_pct = soup.find(attrs={'id': 'div_'+team})

    for row in win_pct.findAll("tr"):
        if len(row.find('th', attrs={'data-stat': 'season'}).text) != 7:
            continue
        if row.find('th', attrs={'data-stat': 'season'}).text == '1979-80':
            break
        data_dict['season'].append(row.find('th', attrs={'data-stat': 'season'}).text)
        data_dict['win_pct'].append(row.find('td', attrs={'data-stat': 'win_loss_pct'}).text)
        data_dict['team'].append(team)


    return data_dict
```

```python
teams = ['ATL','BOS','NJN','CHA','CHI','CLE','DAL','DEN','DET','GSW','HOU','IND',
        'LAC','LAL','MEM','MIA','MIL','MIN','NOH','NYK','OKC','ORL','PHI','PHO',
        'POR','SAC','SAS','TOR','UTA','WAS']
new_data = defaultdict(list)

for team in teams:
    cur_season_dict = work_player_profile(team)
    for key in cur_season_dict:
        new_data[key].extend(cur_season_dict[key])
```

```python
df = pd.DataFrame.from_dict(new_data, orient='index')
```

```python
data_frame = df.transpose()
```

```python
import re
def season_type(row):
    split = str(row['season']).split('-')
    season = split[0][0:2]+split[1]
    return season
data_frame['season'] = data_frame.apply(season_type,axis=1)
```
