```python
import time
import re
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
```

```python
def work_player_profile(season):
    url = 'https://www.basketball-reference.com/leagues/NBA_'+str(season)+'_per_game.html'
    res = requests.get(url)
    soup = BeautifulSoup(res.text)

    print(f"Current season: {season}")

    data_dict = defaultdict(list)

    per_game = soup.find(attrs={'id': 'div_per_game_stats'})
    
    print(f"\tCurrent data: per_game")
    
    index = 0
    for row in per_game.findAll("tr"):
        if 'class' in row.attrs and row.attrs['class'] == ['full_table'] :
            data_dict['player'].append(row.find('td', attrs={'data-stat': 'player'}).text)
            data_dict['pos'].append(row.find('td', attrs={'data-stat': 'pos'}).text)
            data_dict['age'].append(row.find('td', attrs={'data-stat': 'age'}).text)
            data_dict['team'].append(row.find('td', attrs={'data-stat': 'team_id'}).text)
            data_dict['g'].append(row.find('td', attrs={'data-stat': 'g'}).text)
            data_dict['mp'].append(row.find('td', attrs={'data-stat': 'mp_per_g'}).text)
            data_dict['fg'].append(row.find('td', attrs={'data-stat': 'fg_per_g'}).text)
            data_dict['fga'].append(row.find('td', attrs={'data-stat': 'fga_per_g'}).text)
            data_dict['fg%'].append(row.find('td', attrs={'data-stat': 'fg_pct'}).text)
            data_dict['fg3'].append(row.find('td', attrs={'data-stat': 'fg3_per_g'}).text)
            data_dict['fg3a'].append(row.find('td', attrs={'data-stat': 'fg3a_per_g'}).text)
            data_dict['3p%'].append(row.find('td', attrs={'data-stat': 'fg3_pct'}).text)
            data_dict['efg'].append(row.find('td', attrs={'data-stat': 'efg_pct'}).text)
            data_dict['ft'].append(row.find('td', attrs={'data-stat': 'ft_per_g'}).text)
            data_dict['fta'].append(row.find('td', attrs={'data-stat': 'fta_per_g'}).text)
            data_dict['ft%'].append(row.find('td', attrs={'data-stat': 'ft_pct'}).text)
            data_dict['trb'].append(row.find('td', attrs={'data-stat': 'trb_per_g'}).text)
            data_dict['ast'].append(row.find('td', attrs={'data-stat': 'ast_per_g'}).text)
            data_dict['stl'].append(row.find('td', attrs={'data-stat': 'stl_per_g'}).text)
            data_dict['blk'].append(row.find('td', attrs={'data-stat': 'blk_per_g'}).text)
            data_dict['tov'].append(row.find('td', attrs={'data-stat': 'tov_per_g'}).text)
            data_dict['pf'].append(row.find('td', attrs={'data-stat': 'pf_per_g'}).text)
            data_dict['pts'].append(row.find('td', attrs={'data-stat': 'pts_per_g'}).text)
            '''if row.find('td', attrs={'data-stat': 'team_id'}).text != 'TOT':
                base = "https://www.basketball-reference.com"
                link = row.find('td', attrs={'data-stat': 'team_id'}).find("a")['href']
                url = base + link
                time.sleep(0.2)
                soup = BeautifulSoup(requests.get(url).text)
                for item in soup.findAll("p"):
                    if "Record" in item.text:
                        record = re.findall("\d+\-\d+", item.text)[0]
                        splitted = record.split("-")
                        data_dict['win_pct'] = float(splitted[0]) / (float(splitted[1]) + float(splitted[0]))
            else:
                data_dict['win_pct'] = None'''
        index += 1
    
    time.sleep(1)
    
    url = 'https://www.basketball-reference.com/leagues/NBA_'+str(season)+'_advanced.html'
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    

    advanced = soup.find(attrs={'id': 'div_advanced_stats'})
    
    print(f"\tCurrent data: advanced")
    
    index = 0
    for row in advanced.findAll("tr"):
        if 'class' in row.attrs and row.attrs['class'] == ['full_table']:
            data_dict['per'].append(row.find('td', attrs={'data-stat': 'per'}).text)
            data_dict['ts'].append(row.find('td', attrs={'data-stat': 'ts_pct'}).text)
            data_dict['usg'].append(row.find('td', attrs={'data-stat': 'usg_pct'}).text)
            data_dict['ws'].append(row.find('td', attrs={'data-stat': 'ws'}).text)
            data_dict['ws/48'].append(row.find('td', attrs={'data-stat': 'ws_per_48'}).text)
            data_dict['bpm'].append(row.find('td', attrs={'data-stat': 'bpm'}).text)
            data_dict['vorp'].append(row.find('td', attrs={'data-stat': 'vorp'}).text)
            data_dict['season'].append(season)
        index += 1

    return data_dict
```

```python
seasons = range(1981,2021)
new_data = defaultdict(list)

for season in seasons:
    cur_season_dict = work_player_profile(season)
    for key in cur_season_dict:
        new_data[key].extend(cur_season_dict[key])
```

```python
df = pd.DataFrame.from_dict(new_data, orient='index')
```

```python
data_frame = df.transpose()
```
