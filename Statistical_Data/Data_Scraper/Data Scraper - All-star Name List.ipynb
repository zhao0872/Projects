```python
import time
import re
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
```

```python
def work_player_profile(season):
    url = 'https://www.basketball-reference.com/allstar/NBA_'+str(season)+'.html'
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    print(season)
    
    all_stars = soup.findAll("tbody")
    data_dict = defaultdict(list)
    for row in all_stars:
        for player in row.find_all('tr'):
            data_dict['season'].append(str(season))
            data_dict['player'].append(player.find('th').text)
    
    return data_dict
```

```python
seasons = range(1980,2021)
new_data = defaultdict(list)

for season in seasons:
    cur_season_dict = work_player_profile(season)
    for key in cur_season_dict:
        new_data[key].extend(cur_season_dict[key])
```

```python
df = pd.DataFrame.from_dict(new_data, orient='index')
```

```python
data_frame = df.transpose()
```
